# Building the model

```python
#Splittig the data into input and output
X=df.drop(['Exited'],axis=1)
Y=df['Exited']
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.preprocessing import LabelEncoder

#Applying the Label Encoder to the required columns
le = LabelEncoder()
X['Gender'] = le.fit_transform(X['Gender'])
X['Geography'] = le.fit_transform(X['Geography'])

X.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

#Splitting the data into training and testing set
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25)

#Feature Scaling
Scaler = StandardScaler()
X_train = Scaler.fit_transform(X_train)
X_test = Scaler.transform(X_test)
```

```python
len(X_train)
```

```python
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense

# Initializing the ANN model
model = Sequential()

# Adding the input layer and the first hidden layer
model.add(Dense(units=128, activation='relu', input_dim=X_train.shape[1]))

# Adding the second hidden layer
model.add(Dense(units=64, activation='relu'))

# Adding the third hidden layer
model.add(Dense(units=32, activation='relu'))

# Adding the output layer with a single neuron (corresponding to the churn prediction)
model.add(Dense(units=1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

#Fitting the model on the training dataset
epochs = 100
batch_size = 32
history = model.fit(X_train, Y_train.to_numpy(), epochs=epochs, batch_size=batch_size, validation_data=(X_test, Y_test))

# Extracting the training loss and validation loss from the history object
training_loss = history.history['loss']
validation_loss = history.history['val_loss']
training_accuracy = history.history['accuracy']
validation_accuracy = history.history['val_accuracy']

# Printing the loss for each epoch during training
for epoch in range(epochs):
    print(f"Epoch {epoch + 1}/{epochs} - Loss: {training_loss[epoch]:.4f}")

# A plot the training and validation loss
plt.plot(np.arange(1, epochs + 1), training_loss, label='Training Loss')
plt.plot(np.arange(1, epochs + 1), validation_loss, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Evaluation of the model accuracy on the test data
_, accuracy = model.evaluate(X_test, Y_test)
print(f"Accuracy on Test Data: {accuracy:.4f}")
```

```python
plt.plot(np.arange(1, epochs + 1), training_accuracy, label='Training Accuracy')
plt.plot(np.arange(1, epochs + 1), validation_accuracy, label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

Surprisingly, the training loss is being minimized as the epochs progress, however, the validation loss keeps increasing. Despite this, the accuracy is very good on the testing set with a 0.81 score. This discrepancy might indicate overfitting, and the fact that the accuracy on the testing set is still relatively good suggests that the model might be learning noise or peculiarities from the training data that don't generalize well.

To solve this issue, regulization techniques such as dropout might be so useful to adress the neural network over-fitting problem. Dropout randomly deactivates a fraction of neurons during training, which can help prevent the network from relying too much on specific neurons and features.

The code will remain the same as the previous one but will include the addition of dropout.

```python
from keras.layers import Dropout

model = Sequential()

model.add(Dense(units=128, activation='relu', input_dim=X_train.shape[1]))
model.add(Dropout(0.5))  # Adding dropout with a dropout rate of 0.5 (for each hidden layer)

model.add(Dense(units=64, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(units=32, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(units=1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

epochs = 100
batch_size = 32
history = model.fit(X_train, Y_train.to_numpy(), epochs=epochs, batch_size=batch_size, validation_data=(X_test, Y_test))

training_loss = history.history['loss']
validation_loss = history.history['val_loss']

plt.plot(np.arange(1, epochs + 1), training_loss, label='Training Loss')
plt.plot(np.arange(1, epochs + 1), validation_loss, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

_, accuracy = model.evaluate(X_test, Y_test)
print(f"Accuracy on Test Data: {accuracy:.4f}")
```

No more overfitting, and the model's behavior has improved! The accuracy on the testing set has increased to 0.86.

While the accuracy is high, when working with datasets, it's crucial to select evaluation metrics that offer a comprehensive insight into the model's performance across both majority and minority classes. To achieve this, it's prudent to incorporate the recall score as a metric as well. This approach allows us to find a balance between accuracy and recall, thereby addressing churn cases and churn predictions simultaneously.

Recall is the ratio of correctly predicted positive observations (true positives) to the total actual positives (true positives + false negatives). High recall indicates that the model is able to correctly identify most of the actual positive instances. This metric is important the goal is to capture as many positive cases as possible, especially when dealing with rare events like churn (churners are only 20% of total customers).



```python
from sklearn.metrics import recall_score

Y_pred = model.predict(X_test)
Y_pred_class = np.round(Y_pred)
recall = recall_score(Y_test, Y_pred_class)
print(f"Recall Score: {recall:.4f}")
```

The recall score is low, the model is correctly identifying only 45% of the churn cases, while it's missing 55% of them.This can be due to various factors, but in this particular situation, it's most likely due to imbalanced classes.

To adress this, Over-sampling can be applied to the dataset. For this purpose, the Synthetic Minority Over-sampling Technique (SMOTE) can be used.

```python
from imblearn.over_sampling import SMOTE

# Applying SMOTE to the training data
smote = SMOTE(random_state=42)
X_train_resampled, Y_train_resampled = smote.fit_resample(X_train, Y_train)

model = Sequential()

model.add(Dense(units=128, activation='relu', input_dim=X_train_resampled.shape[1]))
model.add(Dropout(0.5))

model.add(Dense(units=64, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(units=32, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(units=1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

epochs = 100
batch_size = 32
history = model.fit(X_train_resampled, Y_train_resampled, epochs=epochs, batch_size=batch_size, validation_data=(X_test, Y_test))

_, accuracy = model.evaluate(X_test, Y_test)
print(f"Accuracy on Test Data: {accuracy:.4f}")

y_pred = model.predict(X_test)
y_pred_class = np.round(y_pred)
recall = recall_score(Y_test, y_pred_class)
print(f"Recall Score: {recall:.4f}")
```

The recall score has improved! Although the accuracy experienced a slight drop, it isn't a concern since we are aiming for a trade-off between the two metrics. We can experiment with adjusting the hyperparameters such as epochs, batch size, and learning rate to observe their impact on the outcomes.

```python
from keras.optimizers import Adam


smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, Y_train)

model = Sequential()

model.add(Dense(units=128, activation='relu', input_dim=X_train_resampled.shape[1]))
model.add(Dropout(0.5))

model.add(Dense(units=64, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(units=32, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(units=1, activation='sigmoid'))


optimizer = Adam(learning_rate=0.001)  #To test with different learning rate values
model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])

#To test with different values
epochs = 200
batch_size = 64

history = model.fit(X_train_resampled, y_train_resampled, epochs=epochs, batch_size=batch_size, validation_data=(X_test, Y_test))

_, accuracy = model.evaluate(X_test, Y_test)
print(f"Accuracy on Test Data: {accuracy:.4f}")

y_pred = model.predict(X_test)
y_pred_class = np.round(y_pred)
recall = recall_score(Y_test, y_pred_class)
print(f"Recall Score: {recall:.4f}")
```

Let's further enhance our results by incorporating Ensemble methods alongside the neural network. By combining the predictions from both models, we can make more informed and accurate final predictions.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

smote = SMOTE(random_state=42)
X_train_resampled, Y_train_resampled = smote.fit_resample(X_train, Y_train)

model = Sequential()

model.add(Dense(units=128, activation='relu', input_dim=X_train_resampled.shape[1]))
model.add(Dropout(0.5))

model.add(Dense(units=64, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(units=32, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(units=1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

epochs = 100
batch_size = 32
history = model.fit(X_train_resampled, Y_train_resampled, epochs=epochs, batch_size=batch_size, validation_data=(X_test, Y_test))

rf_classifier = RandomForestClassifier(random_state=42)

# Training the RandomForest classifier on the original imbalanced data
rf_classifier.fit(X_train, Y_train)

# Combining predictions from both models
nn_pred = model.predict(X_test)
rf_pred = rf_classifier.predict(X_test)

# Flattening the nn_pred array and applying threshold
nn_pred_flat = nn_pred.ravel() > 0.5

# Ensemble prediction: If either model predicts 1, the ensemble predicts 1
ensemble_pred_binary = np.logical_or(nn_pred_flat, rf_pred == 1).astype(int)
ensemble_recall = recall_score(Y_test, ensemble_pred_binary, average='binary')

# Ensemble accuracy
ensemble_accuracy = accuracy_score(Y_test, ensemble_pred_binary)
print(f"Ensemble Accuracy: {ensemble_accuracy:.4f}")

# Ensemble Recall
print(f"Ensemble Recall Score: {ensemble_recall:.4f}")
```

An accuracy of 0.8 and a recall of 0.68 strike a balance between overall correctness and the model's ability to capture positive cases. This combination is great because it indicates that the model is performing well across different dimensions. If we want to improve the model further more, we need to answer the following question that can only be ansewered by domain experts stakeholders: Are missed churn cases more costly than incorrect churn predictions?
